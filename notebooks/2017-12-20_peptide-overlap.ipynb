```python
# perfect, this is the correct number of rows
# now I want to remove the header, and the tryptic ends 
#'K.' and '.Q' in the first peptide for example, are not a part of the peptide used to make the protein ID)

!sed -i -e "1d" /Users/meganduffy/Documents/git-repos/2012-etnp/data/etnp-2012-129_20171219_peps.csv
```

```python
!head /Users/meganduffy/Documents/git-repos/2012-etnp/data/etnp-2012-129_20171219_peps.csv
```

```python
# first I want to get rid of the modification using sed - to clean things up
# it shouldn't matter if we get rid of them or not and I think it will simplfy the data manipulation

# remove '(+15.99)' and '(+57.02)' from sequences
# rewrite to more descriptive file name (that includes 'peaks')
!cat /Users/meganduffy/Documents/git-repos/2012-etnp/data/etnp-2012-129_20171219_peps.csv \
| sed 's/(+15.99)//g' \
| sed 's/(+57.02)//g'\
> /Users/meganduffy/Documents/git-repos/2012-etnp/data/etnp-2012-129_20171219_nomodpeps.csv
```

```python
!head /Users/meganduffy/Documents/git-repos/2012-etnp/data/etnp-2012-129_20171219_nomodpeps.csv
```

```python
# that worked
# now I have a dilemma, because the modified residues at the begininng of the peptide have 
# no given tryptic cleavage residue that's denoted with a '.'

# I've decided to go ahead and (for now) proceed with removing the first 2 and last 2 characters of each line
# then my plan is to go back and write an R script or python script to deal with inconsistencies in lines
# as a first pass, identical peptides would still be ID'd in this manner - we just might have a few false positives

# using sed to trim first 2, last 2 characters:

!cat /Users/meganduffy/Documents/git-repos/2012-etnp/data/etnp-2012-129_20171219_nomodpeps.csv \
| sed 's/^..//' \
| sed 's/..$//'\
> /Users/meganduffy/Documents/git-repos/2012-etnp/data/etnp-2012-129_20171219_nomodpeps_trim.csv
```
